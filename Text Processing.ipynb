```python
import pandas as pd
import numpy as np

from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
df = pd.read_excel("notes_sample.xlsx")
```

```python
del df['customerid']
```

```python
df = df.groupby(['noteid']).sum()
```

```python
df['noteid'] = df.index
```

```python
df = df.set_index('noteid')
```

```python
df = df.dropna()
```

```python
df.to_csv("dropped na- notes.csv")
```
